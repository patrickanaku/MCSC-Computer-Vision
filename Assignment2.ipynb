Import the necessary libraries forworking with image and arrays

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Load the sample image of a dog from google drive folder

```python
# Load the image
image = cv2.imread(image_folder, cv2.IMREAD_GRAYSCALE)  # Convert to grayscale
```

Display the gray scale version of the image

```python
# Show the image
plt.figure(figsize=(6, 6))
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")
plt.show()
```

PErform SIFT operations and display resultant image

```python
# Initialize SIFT detector
sift = cv2.SIFT_create()

# Detect keypoints and compute descriptors
keypoints_sift, descriptors_sift = sift.detectAndCompute(image, None)

# Draw keypoints on the image
sift_image = cv2.drawKeypoints(image, keypoints_sift, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Show the SIFT keypoints image
plt.figure(figsize=(6, 6))
plt.imshow(sift_image, cmap='gray')
plt.title("SIFT Keypoints")
plt.axis("off")
plt.show()
```

PErform ORB operations and display resultant image

```python
# Initialize ORB detector
orb = cv2.ORB_create()

# Detect keypoints and compute descriptors
keypoints_orb, descriptors_orb = orb.detectAndCompute(image, None)

# Draw keypoints on the image
orb_image = cv2.drawKeypoints(image, keypoints_orb, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Show the ORB keypoints image
plt.figure(figsize=(6, 6))
plt.imshow(orb_image, cmap='gray')
plt.title("ORB Keypoints")
plt.axis("off")
plt.show()
```
